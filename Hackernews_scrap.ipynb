```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}
#checking status code
# webpage =requests.get('https://news.ycombinator.com/',headers = headers).status_code
#200 is succesful
webpage =requests.get('https://news.ycombinator.com/',headers = headers).text

```

```python
#creating the Beautiful soup variable in order to use the methods
soup = BeautifulSoup(webpage,'lxml')
```

```python
news_link = soup.find_all('span',class_='titleline')
```

```python
links = []
titles = []
for i in news_link:
    links.append(i.find('a').get('href'))
    titles.append(i.find('a').text.strip())
```

```python
titles
```

```python
d = {'News_title':titles, 'Link':links}
df = pd.DataFrame(d)
```

```python
final = pd.DataFrame()
for j in range(1,11):
    url = 'https://news.ycombinator.com/'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}
    webpage =requests.get(url,headers = headers).text
    soup = BeautifulSoup(webpage,'lxml')
    news_link = soup.find_all('span',class_='titleline')
    links = []
    titles = []
    for i in news_link:
        links.append(i.find('a').get('href'))
        titles.append(i.find('a').text.strip())
    d = {'News_title':titles, 'Link':links}
    df = pd.DataFrame(d)
    final = final.append(df)
```

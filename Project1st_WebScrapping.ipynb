```python
final = pd.DataFrame()

for j in range(1,21):
    url = 'http://books.toscrape.com/'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}
    webpage = requests.get(url,headers = headers).text
    soup = BeautifulSoup(webpage,'lxml')
    books = soup.find_all('article',class_='product_pod')
    books_title = []
    rating = []
    prices = []
    for i in books:
        books_title.append(i.find('h3').text.strip())
        rating.append(i.find('p').get('class'))
        prices.append(i.find('div',class_= 'product_price').text.strip())
        f_prices = []
    stocks = []
    for i in prices:
        f_prices.append(i.split('\n')[0].replace('Â',''))
        stocks.append(i.split('\n')[4].replace(' ',''))
    d = {'Books_title': books_title, 'Rating':rating,'Prices':f_prices, 'Stock_availabile':stocks}
    df = pd.DataFrame(d)   
    final = final.append(df)
    
```
